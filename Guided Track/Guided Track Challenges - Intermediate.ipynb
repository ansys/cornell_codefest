> Note: If you want to continue working entirely in the notebook instead of switching to a local script. This is possible! Normally when you use `argparse` you have to call the following code (*somewhere*) to parse the input arguments:
>
> ```python
> parser.parse_args()
> ```
> This then goes and gets the python script arguments when it was run and you can use your CLI. 
> However, you can *also* provide a list of strings to parse instead!
>
> ```python
> parser.parse_args(['--file_size', 'test_file.txt'])
> ```
> And the parser will parse the string you provide instead of the scripts arguments. This essentially lets you achieve the same effect as calling the script on the command line!
> 
> If you choose to do this your notebook should ask for inputs to be parsed using the `input` function and each scenario above should still work. Although, you will necessarily have to omit the `python` from each string. An example of this in action is included in the cell below, which you can play around with.
>
> E.g. the input for **A** would be `your_script.py --print_line test_file.txt`.


````python
"""
This cell contains a very basic CLI for a jupyter notebook. To test it, run this cell and enter

--show "Hello World!"

in the prompt.

Try it again with

--help

as well! Note if the cell exits with SystemExit: 0, do not worry, this is fine! 
Jupyter may get a bit confused by argparse, however, it's not a problem.
"""
import argparse
import shlex

# Create the parser
parser = argparse.ArgumentParser()

# Add a flag/argument to the parser
parser.add_argument('--show', help='prints to screen whichever argument is provided')

# Take the user's input in lieu of the command line inputs
user_input = input('Enter CLI flags and args:')

# Our input is a string and it needs to be a list of arguments for parse_args to understand it
# We use the builtin library shlex to split the string by space, whilst preserving substrings in quotes
user_input = shlex.split(user_input)

# parse the arguments we've provided
args = parser.parse_args(user_input)

# print the value supplied for the --show flag
print(args.show)
````
